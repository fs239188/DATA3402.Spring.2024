```python
#Qusetion 3 sol
def constant(n, m, c):
    return Matrix([[c] * m for _ in range(n)])

def zeros(n, m):
    return constant(n, m, 0)

def ones(n, m):
    return constant(n, m, 1)

def eye(n):
    answer = zeros(n, n)
    for i in range(n):
        answer[i, i] = 1
    return answer
```

```python
#Question 5 sol
class Matrix:
    def __init__(self, *args):
        if len(args) == 2 and isinstance(args[0], int) and isinstance(args[1], int):
            self.n = args[0]
            self.m = args[1]
            self.data = [[0] * self.m for _ in range(self.n)]
        elif len(args) == 1 and isinstance(args[0], list):
            rows = len(args[0])
            cols = len(args[0][0])
            for row in args[0]:
                if len(row) != cols:
                    raise ValueError("Rows need to have the same no of cols")
            self.n = rows
            self.m = cols
            self.data = args[0]
        else:
            raise ValueError("Error")

    def __getitem__(self, key):
        if isinstance(key, tuple) and len(key) == 2:
            i, j = key
            return self.data[i][j]
        elif isinstance(key, int):
            return self.data[key]
        elif isinstance(key, slice):
            return [self.data[i] for i in range(*key.indices(self.n))]
        else:
            raise TypeError("Indexing Issue")

    def __setitem__(self, key, value):
        if isinstance(key, tuple) and len(key) == 2:
            i, j = key
            self.data[i][j] = value
        elif isinstance(key, int):
            self.data[key] = value
        else:
            raise TypeError("Indexing Issue")

    def __eq__(self, other):
        if isinstance(other, Matrix):
            return self.data == other.data
        elif isinstance(other, list):
            return self.data == other
        else:
            return False

    def __str__(self):
        return '\n'.join([' '.join(map(str, row)) for row in self.data])

    def shape(self):
        return self.n, self.m

    def transpose(self):
        the_transposed_data = [[self.data[j][i] for j in range(self.n)] for i in range(self.m)]
        return Matrix(the_transposed_data)

    def row(self, n):
        return Matrix([self.data[n]])

    def column(self, n):
        return Matrix([[self.data[i][n]] for i in range(self.n)])

    def to_list(self):
        return self.data

    def block(self, n_0, n_1, m_0, m_1):
        the_block_data = [row[m_0:m_1] for row in self.data[n_0:n_1]]
        return Matrix(the_block_data)

    def scalarmul(self, c):
        return Matrix([[c * self.data[i][j] for j in range(self.m)] for i in range(self.n)])

    def add(self, other):
        if self.n != other.n or self.m != other.m:
            raise ValueError("The matrices need tohave the same dimensions to add")
        return Matrix([[self.data[i][j] + other.data[i][j] for j in range(self.m)] for i in range(self.n)])

    def sub(self, other):
        if self.n != other.n or self.m != other.m:
            raise ValueError("The matrices need tohave the same dimensions to subtract")
        return Matrix([[self.data[i][j] - other.data[i][j] for j in range(self.m)] for i in range(self.n)])

    def mat_mult(self, other):
        if self.m != other.n:
            raise ValueError("The no of cols in first matrix need to equal to the no of rows in second matrix to multiply")
        answer = [[sum(self.data[i][k] * other.data[k][j] for k in range(self.m)) for j in range(other.m)] for i in range(self.n)]
        return Matrix(answer)

    def element_mult(self, other):
        if self.n != other.n or self.m != other.m:
            raise ValueError("The matrices need to have the same dimension")
        return Matrix([[self.data[i][j] * other.data[i][j] for j in range(self.m)] for i in range(self.n)])

    def equals(self, other):
        return self == other
    
    def __mul__(self, other):    #overloading operators
        if isinstance(other, (int, float)):
            return self.scalarmul(other)
        elif isinstance(other, Matrix):
            return self.mat_mult(other)
        else:
            raise TypeError("Cant work for *: '{}' and '{}'".format(type(self).__name__, type(other).__name__))

    def __rmul__(self, other):
        return self.__mul__(other)

    def __add__(self, other):
        if isinstance(other, Matrix):
            return self.add(other)
        else:
            raise TypeError("Cant work for + : '{}' and '{}'".format(type(self).__name__, type(other).__name__))

    def __sub__(self, other):
        if isinstance(other, Matrix):
            return self.sub(other)
        else:
            raise TypeError("Cant work for - : '{}' and '{}'".format(type(self).__name__, type(other).__name__))

    def __eq__(self, other):
        if isinstance(other, Matrix):
            return self.equals(other)
        else:
            raise False

    def __ne__(self, other):
        return not self == other
```
